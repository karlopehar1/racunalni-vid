```python
### 3. zadatak
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/goldhill.bmp', cv2.IMREAD_GRAYSCALE)

ret, thresh_cv = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
thresh_np = np.where(img > 128, 255, 0).astype(np.uint8)

plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1,3,2)
plt.imshow(thresh_cv, cmap='gray')
plt.title('Thresholding OpenCV')

plt.subplot(1,3,3)
plt.imshow(thresh_np, cmap='gray')
plt.title('Thresholding NumPy')

plt.show()
```

```python
### 4. zadatak
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images/goldhill.bmp', cv2.IMREAD_GRAYSCALE)

ret_otsu, otsu_thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print("Optimalni prag prema Otsu metodi:", ret_otsu)

plt.figure(figsize=(6,6))
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu Binarizacija")
plt.show()
```

```python
### 5. zadatak
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images/goldhill.bmp', cv2.IMREAD_GRAYSCALE)
ret_otsu, otsu_thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

img_thresholded = cv2.bitwise_and(img, img, mask=otsu_thresh)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Originalna slika")
plt.subplot(1,2,2)
plt.imshow(img_thresholded, cmap='gray')
plt.title("Slika nakon primjene Otsu maske")
plt.show()
```

```python
### Tasks
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/binary_image.png', 0)

rect = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))
ellip = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (7,7))

closing_rect = cv2.morphologyEx(img, cv2.MORPH_CLOSE, rect)
closing_ellip = cv2.morphologyEx(img, cv2.MORPH_CLOSE, ellip)
closing_cross = cv2.morphologyEx(img, cv2.MORPH_CLOSE, cross)

plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.imshow(img, cmap='gray')
plt.title('Originalna')
plt.subplot(1,4,2)
plt.imshow(closing_rect, cmap='gray')
plt.title('Closing - Rectangle')
plt.subplot(1,4,3)
plt.imshow(closing_ellip, cmap='gray')
plt.title('Closing - Ellipse')
plt.subplot(1,4,4)
plt.imshow(closing_cross, cmap='gray')
plt.title('Closing - Cross')
plt.show()
```
